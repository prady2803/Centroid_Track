```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-


import cv2
import argparse
from operator import xor



def setup_trackbars(range_filter):
    cv2.namedWindow("Trackbars", 0)

    for i in ["MIN", "MAX"]:
        v = 0 if i == "MIN" else 255

        for j in range_filter:
            cv2.createTrackbar("%s_%s" % (j, i), "Trackbars", v, 255, on_trackbar)

def get_trackbar_values(range_filter):
    values = []

    for i in ["MIN", "MAX"]:
        for j in range_filter:
            v = cv2.getTrackbarPos("%s_%s" % (j, i), "Trackbars")
            values.append(v)

    return values

def on_trackbar(val):
    pass

def main():
   
    vid = cv2.VideoCapture(1)
    range_filter = "HSV"
    setup_trackbars(range_filter)
    #ret,frame = vid.read()
    
    while True:
        #vid = cv2.VideoCapture(0)
        ret,frame = vid.read()
#         height,width,channels = frame.shape
#         height = int(height)
#         width = int(width)
#         channels = int(channels)
#         #descentre = 60
        

#         aux1 = int(((height)/2)- 160)
#         aux2 = int(((height)/2) - 140)

#         crop_img = frame[aux1:aux2][1:width]
        crop_img = frame

        frame_to_thresh = cv2.cvtColor(crop_img, cv2.COLOR_BGR2HSV)

        v1_min, v2_min, v3_min, v1_max, v2_max, v3_max = get_trackbar_values(range_filter)

        thresh = cv2.inRange(frame_to_thresh, (v1_min, v2_min, v3_min), (v1_max, v2_max, v3_max))

        preview = cv2.bitwise_and(crop_img, crop_img, mask=thresh)
        cv2.imshow("Preview", preview)

        if cv2.waitKey(1) & 0xFF is ord('q'):
            break
    
    cv2.destroyAllWindows()

if __name__ == '__main__':
    main()
```
