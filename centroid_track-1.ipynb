```python
#!/usr/bin/env python

import cv2
import numpy as np
import snap7
from snap7.util import*
from snap7.types import*
import snap7.client as c

vid = cv2.VideoCapture(1)
plc = c.Client()
plc.connect('192.168.4.252',0,1)
if plc.get_connected():
    print("Connected Successfully")

while (True):
    ret,frame = vid.read()
    height, width, channels = frame.shape
    descentre = 160
    rows_to_watch = 80
    crop_img = frame[int((height)/2+descentre):int((height)/2+(descentre+rows_to_watch))][1:width]

    # Convert from RGB to HSV
    hsv = cv2.cvtColor(crop_img, cv2.COLOR_BGR2HSV)
        
        
    lower_yellow = np.array([0,110,48])
    upper_yellow = np.array([180,255,255])

    # Threshold the HSV image to get only yellow colors
    mask = cv2.inRange(hsv, lower_yellow, upper_yellow)

    # Bitwise-AND mask and original image
    res = cv2.bitwise_and(crop_img,crop_img, mask= mask)

    contours, _, = cv2.findContours(mask, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_TC89_L1)
    print("Number of centroids==>"+str(len(contours)))
    desired_contours = []
    centres = []
    cent_x =[]
    for contour in contours:
        if (cv2.contourArea(contour) > 3000):
            desired_contours.append(contour)
    print("Number of Desired centroids==>"+str(len(desired_contours)))
    for i in range(len(desired_contours)):
        
        moments = cv2.moments(desired_contours[i])
        
        
        try:
            centres.append((int(moments['m10']/moments['m00']), int(moments['m01']/moments['m00'])))
            cv2.circle(res, centres[-1], 10, (0, 255, 0), -1)
            cv2.circle(res, centres[-1], 5,(0,0,255),-1)
        except ZeroDivisionError:
            pass
    print(centres)
    for i in centres:
        cent_x.append(i[0])
    print(cent_x)
    Cx = [0,0,0]
    Cx = cent_x[::-1]
    for i in range(3- len(Cx)):
        Cx.append(0)
    print (Cx)
    plc.write_area(Areas.DB, 2, 0, Cx[0].to_bytes(2,"big"))
    plc.write_area(Areas.DB, 2, 2, Cx[1].to_bytes(2,"big"))
    plc.write_area(Areas.DB, 2, 4, Cx[2].to_bytes(2,"big"))
    
    
       #Select the right centroid
    # [(542, 39), (136, 46)], (x, y)
#     most_right_centroid_index = 0
#     index = 0
#     max_x_value = 0
#     for candidate in centres:
#         # Retrieve the cx value
#         cx = candidate[0]
#         # Get the Cx more to the right
#         if cx >= max_x_value:
#             max_x_value = cx
#             most_right_centroid_index = index
#         index += 1


#     try:
#         cx = centres[most_right_centroid_index][0]
#         cy = centres[most_right_centroid_index][1]
        
#     except:
#         cy, cx = height/2, width/2

    # Draw the centroid in the resulting image
    # cv2.circle(img, center, radius, color[, thickness[, lineType[, shift]]]) 
    #cv2.circle(res,(int(cx), int(cy)), 5,(0,0,255),-1)



    cv2.imshow("Original", frame)
    #cv2.imshow("HSV", hsv)
    #cv2.imshow("MASK", mask)
    cv2.imshow("RES", res)

    if cv2.waitKey(1) & 0xFF is ord('q'):
            break
    
cv2.destroyAllWindows()
   
```
